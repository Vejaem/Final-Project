```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("CloudETL").config("spark.driver.extraClassPath","/content/postgresql-42.2.18.jar").getOrCreate()
```

```python
# Read in data from S3 Buckets
# Read in data from S3 Buckets
from pyspark import SparkFiles
url="https://markdatabucket.s3.amazonaws.com/cbb.csv"

spark.sparkContext.addFile(url)
spark_data_df = spark.read.csv(SparkFiles.get("cbb.csv"), sep=",", header=True, inferSchema=True)

# need to convert spark dataframe to pandas dataframe
cbb_df = spark_data_df.toPandas()

# Show DataFrame
cbb_df.head()
```

```python
#filtering down to one year of data, using dot notation and a conditional statement "2015"
cbb_2015_df = cbb_df[cbb_df.YEAR == 2015]
cbb_2015_df.head()
```

```python
# duplicate postseason column to later use as binary option
cbb_2015_df['POSTSEASON_BINARY'] = cbb_2015_df['POSTSEASON']
cbb_2015_df.head()
```

```python
cbb_2015_df['POSTSEASON_BINARY'] = cbb_2015_df['POSTSEASON']
cbb_2015_df.POSTSEASON_BINARY[cbb_2015_df.POSTSEASON_BINARY == "NA"] = 0
cbb_2015_df.POSTSEASON_BINARY[cbb_2015_df.POSTSEASON_BINARY != 0] = 1
cbb_2015_df.head(10)
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = cbb_2015_df.TORD.values.reshape(-1, 1)
y = cbb_2015_df.POSTSEASON_BINARY
y=y.astype('int')
```

```python
# Plot the data
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create the model and fit the model to the data

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python

classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```
