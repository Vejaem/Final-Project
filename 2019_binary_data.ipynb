```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
#read in csv file, utitilizing read.csv method
cbb_df = pd.read_csv("cbb.csv")
cbb_df
```

```python
#filtering down to one year of data, using dot notation and a conditional statement "2019"
cbb_df = cbb_df[cbb_df.YEAR == 2019]
cbb_df
```

```python
# duplicate postseason column to later use as binary option
cbb_df['POSTSEASON_BINARY'] = cbb_df['POSTSEASON']
cbb_df.head()
```

```python
cbb_df['POSTSEASON_BINARY'] = cbb_df['POSTSEASON']
cbb_df.POSTSEASON_BINARY[cbb_df.POSTSEASON_BINARY == "NA"] = 0
cbb_df.POSTSEASON_BINARY[cbb_df.POSTSEASON_BINARY != 0] = 1
cbb_df.head(10)
```
